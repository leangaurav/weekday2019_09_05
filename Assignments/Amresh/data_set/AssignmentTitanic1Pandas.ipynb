```python
#Fill na with mean for fare and age columan
df.age.fillna(df.age.mean(),inplace = True)
df.fare.fillna(df.fare.mean(),inplace= True)
print(df.isna().any())
df.shape
df.dropna(inplace = True)
print(df.shape)

```

```python
print(df.isna().any())
df.count()
```

```python
df.dropna(inplace =True)#delete two rows in which embarked having (NA cell)
df.count()
```

```python
%%time
df.survived.replace({0:'D',1:'A'}, inplace =True)
```

```python
%%time
df.head()
```

```python
r = df.survived.value_counts()
print(r)
print(type(r))
```

```python
print(r.index)
print(r.values)
```

```python
r = df.groupby(['survived','gender'])
print(r)
print(r.groups)
print(r.size())
```

```python
r = df.groupby(['survived','gender'])
print(r.age.min())
print(r.age.mean())
```

```python
df.age.sort_values()
```

```python
r = df.sort_values(by =['age'],ascending = False).head()
print(r)
```

```python
print(r.gender.value_counts())
```

```python
idx = df.survived == 'A'
p = df[idx]
r = p.groupby(['gender'])
print(r.age.max())
print(dir(r.groups))
print(r.groups.keys())
print(type(r))
print(type(r.count()))

```
