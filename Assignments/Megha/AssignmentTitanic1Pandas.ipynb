```python
import numpy as np
import pandas as pd
```

```python
# 1. Read data and describe it

df = pd.read_csv(r'E:\DataScience\weekday2019_09_05\titanic_dataset.csv')
print(df.describe(include = 'all'))
print(df.head())
#Remove Unnecessary columns

(df.count()/1309 ) * 100

df.drop(columns = ['Unnamed: 0' , 'cabin', 'body', 'boat'] , inplace = True)
```

```python
# 2. Find columns with missingdata

print(df.isna().sum())
print()
print("columns with missing data")
for data in df.columns:
    if df[data].isna().any():
        print(data , df[data].isna().sum(), sep = '  ')

# df.columns[df.isna().any()]
```

```python
# 3. Print shape of dataset

df.shape
```

```python
# 4. drop columns with more than 25% missing data

s = df.isna().sum()
print(s)
s1 = s/df.shape[0] * 100 > 25
print("The columns with more than 25% data  missing are :")
print(s[s1])
# df.drop(  columns = s[s1].index , inplace = True)
# df

#Using Thresh

df.dropna(thresh = 1309 * 0.75 , axis = 1 , inplace = True)
df.shape

```

```python
# 5.  drop columns having independent values(which do not affect the survival rate).

print(df.columns)
# df.drop(columns = ['name' , 'ticket'], inplace = True)
print(df.count())
```

```python
# 6. check data types of all columns

df.dtypes
```

```python
# 7. convert price to numeric
# L = list()
# for data in df['fare']:
# #      print(str(df['fare']).replace('$' , ' ').split('\n'))
#     L.append(str(data).replace('$' , ''))
# # print(L)

# df['fare']  = pd.Series(L)


# df['fare'] = [ str(data).replace('$' , '') for data in df['fare'] ]
df

df['fare'].replace(r'\$', '', regex = True , inplace = True)
df['fare'] = df['fare'].astype('float')
df['fare'].dtype
```

```python
# 8. find columns still having missing/na values and also count of missing data

for data in df.columns:
    if df[data].isna().any():
        print(data , df[data].isna().sum())
        
df_mis = pd.DataFrame(df.isna().sum())
df_mis[df_mis[0] > 0]

```

```python
# 9. fill na with mean for fare and age column column.

df.age.fillna(df.age.mean() , inplace = True)
df.fare.fillna(df.fare.mean() , inplace = True)
df.isna().sum()
# df.shape
```

```python
# 10. drop na values for embarked column.

# s = df['embarked'].isna()
# df.drop(index = s[s].index , inplace = True)
# df


df.drop(index = df[df['embarked'].isna()].index , inplace = True)
df

# df.dropna(inplace = True)
```

```python
# find max age male and female who survived


df.groupby(['gender', 'survived']).age.max()

#Using pivot table

df.pivot_table(values = 'age' , index = ['gender', 'survived'] , aggfunc = np.max )
```
