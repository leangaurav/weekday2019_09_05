```python
r = df.gender.value_counts()


plt.figure(figsize = (16,5))

plt.subplot(1,2, 1)
print(r)
print(r.plot(kind = 'bar'))

plt.subplot(1,2, 2)
plt.bar(r.index ,r.values)

for idx in r.index:
    plt.text(idx, r[idx], str( (r[idx], idx) )  )

plt.show()
```

```python
r = df.gender.value_counts()


plt.figure(figsize = (16,5))

# pandas plotting
plt.subplot(1,2, 1)
print(r)
print(r.plot(kind = 'barh'))

# using matplotlib
plt.subplot(1,2, 2)
plt.barh(r.index ,r.values)

plt.show()
```

```python
r = df.gender.value_counts()
percent = r/r.sum() *100
print(percent)

r1=percent.index + ':' + percent.values.astype('int').astype('str')
print(r1)
plt.figure(figsize = (16,5))

plt.subplot(1,2, 1)
print(r)
print(r.plot(kind = 'pie'))



plt.subplot(1,2, 2)
plt.pie(r.values, labels = r.index, autopct='%.1f%%', pctdistance = 0.9)
plt.legend()

plt.show()
```

```python
dff = pd.DataFrame({2001:[10,20,30,23,40], 2009: [40,20,10,5,10], 2019: [40,20,10,5,10]})
print(dff)

plt.bar(dff[2001].index, dff[2001].values, label = '2001')
plt.bar(dff[2009].index, dff[2009].values, bottom = dff[2001].values, label = '2009')
plt.bar(dff[2009].index, dff[2009].values, bottom = dff[2001].values + dff[2009].values, label = '2019')
plt.legend()

plt.show()
```
