## Reversed vs reverse

```python
# create list
a = [1,2,3,4]
r = reversed(a) # gives iterator, doesn't actually reverse
print(r)
print(list(r)) # iterator doesn't make actual change to data
print(a) # list stays as it is
#id's will be different cos the iterator is a wraper object over actual list
print(id(a), id(r)) 
```

```python
# you cannot reverse a tuple, as it is immutable
t = (1,2,3)
#t.reverse()
# but you can get a reversed view of any sequence
print( list(reversed(t) ) )
```

```python
# see this generator to understand how iterators might behave internally
def reversed_gen(seq):
    start = len(seq) - 1
    while start >= 0:
        yield seq[start]
        start -= 1
s = 'abcde'

for c in reversed_gen(s):
    print(c)
```
